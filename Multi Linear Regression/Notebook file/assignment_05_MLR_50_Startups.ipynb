Prepare a prediction model for profit of 50_startups data.
Do transformations for getting better predictions of profit and
make a table containing R^2 value for each prepared model.

R&D Spend -- Research and devolop spend in the past few years
Administration -- spend on administration in the past few years
Marketing Spend -- spend on Marketing in the past few years
State -- states from which data is collected
Profit  -- profit of each state in the past few years.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
import numpy as np 
```

```python
#Read the data
start_u = pd.read_csv("50_Startups.csv")
print(start_u)
```

```python
start_u =start_u.rename({'R&D Spend':'R_DS','Administration':'AD_MS','Marketing Spend':'MK_TS'},axis=1)
```

```python
start_u.info()
```

```python
#check for missing values
start_u.isnull().sum()
```

###**Correlation Matrix**

```python
start_u.corr()
```

###**Scatterplot between variables along with histograms**

```python
#Format the plot background and scatter plots for all the variables
sns.set_style(style='darkgrid')
sns.pairplot(start_u)
```

###**Preparing a model**

```python
#Build model
import statsmodels.formula.api as smf 
model = smf.ols('Profit~R_DS+AD_MS+MK_TS',data=start_u).fit()
```

```python
#Coefficients
model.params
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

###**Regression Models**

```python
ml_AD_MS = smf.ols('Profit~AD_MS',data=start_u).fit()
#t and p-Values
print(ml_AD_MS.tvalues, '\n', ml_AD_MS.pvalues) 
```

```python
ml_MK_TS = smf.ols('Profit~MK_TS',data=start_u).fit()
#t and p-Values
print(ml_MK_TS.tvalues, '\n', ml_MK_TS.pvalues)
```

```python
ml_AD_MK =smf.ols('Profit~AD_MS+MK_TS',data = start_u).fit()  
print(ml_AD_MK.tvalues, '\n', ml_AD_MK.pvalues) 
```

###**Calculating VIF** 

```python
rsq_R_DS =smf.ols('R_DS~AD_MS+MK_TS',data=start_u).fit().rsquared  
vif_R_DS = 1/(1-rsq_R_DS)

rsq_AD_MS = smf.ols('AD_MS~R_DS+MK_TS',data=start_u).fit().rsquared  
vif_AD_MS = 1/(1-rsq_AD_MS)

rsq_MK_TS = smf.ols('MK_TS~R_DS+AD_MS',data=start_u).fit().rsquared  
vif_MK_TS = 1/(1-rsq_MK_TS)

# Storing vif values in a data frame
d1 = {'Variables':['R_DS','AD_MS','MK_TS'],'VIF':[vif_R_DS,vif_AD_MS,vif_MK_TS]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

###**Residual** **Analysis**

Test for Normality of Residuals (Q-Q Plot)

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid>10))
```

###**Residual Plot for Homoscedasticity**

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "MK_TS", fig=fig)
plt.show()
```

###**Model Deletion Diagnostics**

**Detecting Influencers/Outliers**

**Cooks** **Distance**

```python
#Cook's Distance: If Cook's distance > 1, then it's an outlier
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(start_u)), np.round(c, 5))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#Index and value of influencer where c is maximum with the range.
(np.argmax(c),np.max(c))
```

###**Improving the Model**

```python
# Discard the data points which are influencers and reassign the row number (reset_index(drop=True))
start_1=start_u.drop(start_u.index[[49]],axis=0).reset_index(drop=True)
start_1
```

Since the value is <1 , we can stop the diagnostic process and finalize the model

```python
#Check the accuracy of the mode
final_ml_V= smf.ols('Profit~R_DS+AD_MS+MK_TS',data = start_1).fit()
```

```python
(final_ml_V.rsquared,final_ml_V.aic)
```

```python
model_influence_V = final_ml_V.get_influence()
(c_V, _) = model_influence_V.cooks_distance 
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
fig= plt.subplots(figsize=(20,7))
plt.stem(np.round(c_V,3));
plt.xlabel('Row index')
plt.ylabel('Cooks Distance'); 
```
