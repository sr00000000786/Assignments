# **K-Mean Cluster**

```python
from sklearn.cluster import KMeans
```

```python
crime_1 = pd.read_csv("crime_data.csv")
crime_1
```

```python
def norm_func(i):
    x = (i-i.min()) / (i.max() - i.min())
    return (x)  
```

```python
df_norm = norm_func(crime_1.iloc[:,1:])  
```

```python
df_norm.head(10)
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(6, 4))
WCSS = []     #Within cluster sum of square(empty list created)
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()  
```

Cluster using K=5

```python
clf = KMeans(n_clusters=5) #K=5
y_kmeans = clf.fit_predict(df_norm) #Fit and predict on normalize data
```

```python
y_kmeans
#clf.cluster_centers_
clf.labels_ 
```

```python
y_kmeans 
```

```python
clf.cluster_centers_ 
```

```python
clf.inertia_
```

```python
md=pd.Series(y_kmeans)  # converting numpy array into pandas series object 
crime_1['clust']=md # creating a  new column and assigning it to new column 
crime_1
```

```python
crime_1.iloc[:,1:6].groupby(crime_1.clust).mean() #Group by cluster and findout the mean(avg.)
```

```python
plt.style.use(['classic']) 
print(plt.style.available)  
```

```python
crime_1.plot(x="Murder",y ="Assault",c=clf.labels_,kind="scatter",s=30 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans') 
```

# **DB-SCAN model**

```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# Import .csv file and convert it to a DataFrame object
crime_DB = pd.read_csv("crime_data.csv")
crime_DB
```

```python
print(crime_DB.info()) 
```

```python
cdb = crime_DB.drop(['Unnamed: 0'],axis=1) 
```

```python
array=cdb.values
array 
```

```python
stscaler = StandardScaler().fit(array) #Standardized data & converting to Z-score
X = stscaler.transform(array) 
X   
```

```python
from sklearn.neighbors import NearestNeighbors  
```

```python
neigh = NearestNeighbors(n_neighbors=8) #as we have 4 dimension n=2*d
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X) 
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)  
```

```python
dbscan = DBSCAN(eps=1.02, min_samples=5) #take eps from above distance plot
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster']) 
```

```python
cl
pd.set_option("display.max_rows", None) 
```

```python
cl
```

```python
df1 = pd.concat([crime_DB,cl],axis=1)  
df1 
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
df1.plot(x="Murder",y ="Rape",c=dbscan.labels_ ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using DBScan')  
```

```python
dl = dbscan.labels_
```

```python
import sklearn
sklearn.metrics.silhouette_score(X, dl)  
```

```python
from sklearn.cluster import KMeans #Using K-mean for better silhoutee score 
clf = KMeans(n_clusters=4)
y_kmeans = clf.fit_predict(X)  
```

```python
y_kmeans
```

```python
cl1=pd.DataFrame(y_kmeans,columns=['Kcluster']) 
cl1 
```

```python
df2 = pd.concat([df1,cl1],axis=1) 
df2 
```

```python
df2.plot(x="Murder",y ="Rape",c=y_kmeans ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using KMeans') 
```

```python
sklearn.metrics.silhouette_score(X, y_kmeans)
```

###Linkage Metrics:Complete

```python
# Extract the features from the dataset
features = crime[['Murder', 'Assault', 'Rape']]  

# Perform complete linkage clustering
linkage_matrix = linkage(features, method='complete')

# Plot the dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linkage_matrix)
plt.title("Complete Linkage Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

###Linkage Metrics:Ward

```python
# Extract the features from the dataset
features = crime[['Murder', 'Assault', 'Rape']] 

# Perform Ward's linkage clustering
linkage_matrix = linkage(features, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linkage_matrix)
plt.title("Ward's Linkage Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

###Distance Methods:  Euclidean Distance

```python
from scipy.spatial.distance import euclidean
# Extract the features from the dataset
p1 = crime.loc[0, ['Murder', 'Assault', 'Rape']]   
p2 = crime.loc[1, ['Murder', 'Assault', 'Rape']]  

# Calculate the Euclidean distance between the two points
distance = euclidean(p1, p2)

# Print the distance
print("Euclidean distance:", distance)
```

###Distance Methods: cosine distance

```python
from scipy.spatial.distance import cosine
# Extract the features from the dataset
p3 = crime.loc[0, ['Murder', 'Assault', 'Rape']]   
p4 = crime.loc[1, ['Murder', 'Assault', 'Rape']]  

# Calculate the cosine distance between the two points
distance = cosine(p3,p4)

# Print the distance
print("Cosine distance:", distance)
```
