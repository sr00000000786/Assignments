### Calculate the Pearson correlation coefficient matrix


```python
correlation_matrix = book_df.corr(method='pearson')
```

```python
# Example: Find similar users based on the correlation with a specific user
user_id = 276726

# Check if the user ID exists in the correlation matrix
if user_id in correlation_matrix.index:
    similar_users = correlation_matrix[user_id].drop(user_id).sort_values(ascending=False)
    
    # Print the most similar users
    print("Most similar users:")
    print(similar_users)
else:
    print(f"User {user_id} not found in the correlation matrix.")
```

###Calculate the Jaccard Similarity

```python
# Group the data by User_ID and create sets of titles for each user
user_sets = book_df.groupby("UserID")["Tittle"].apply(set)

# Calculate the Jaccard similarity between two users
user1_id = 276729
user2_id = 276729

if user1_id in user_sets.index and user2_id in user_sets.index:
    user1_titles = user_sets[user1_id]
    user2_titles = user_sets[user2_id]
    
    # Calculate the Jaccard similarity
    jaccard_similarity = len(user1_titles.intersection(user2_titles)) / len(user1_titles.union(user2_titles))
    
    print(f"Jaccard similarity between User {user1_id} and User {user2_id}: {jaccard_similarity:.2f}")
else:
    print("Invalid user ID.")
```

###Neighborhood-Based Collaborative Filtering

```python
from sklearn.metrics.pairwise import pairwise_distances
# Create a pivot table with User_ID as rows and Title as columns
pivot_table = book_df.pivot_table(index="UserID", columns="Tittle", values="Rating", fill_value=0)

# Calculate pairwise distances between users using Jaccard similarity
distances = pairwise_distances(pivot_table.values, metric="jaccard")

# Example: Find similar users based on nearest neighbors for a specific user
user_id = 1
k = 5  # Number of nearest neighbors

# Get the indices of k nearest neighbors (excluding the user itself)
user_distances = distances[user_id - 1]
nearest_neighbors = user_distances.argsort()[1:k + 1]

# Print the most similar users
print("Most similar users:")
for neighbor in nearest_neighbors:
    print("User:", neighbor + 1)
```
